```python
def create_thumbnail(input_file_path, thumbnail_path, thumbnail_size):
    with Image.open(input_file_path) as Im:
        Im.thumbnail(thumbnail_size)
        Im.save(thumbnail_path, format="JPEG")
        
create_thumbnail("Downloads/gold.png","gold_thumbnail.jpg",(128, 128))
```
